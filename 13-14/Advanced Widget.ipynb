```python
import tkinter as tk
from tkinter import ttk

app = tk.Tk()
app.geometry("300x600")

listBox = tk.Listbox(app,selectmode=tk.MULTIPLE)
pc_part = ["مانیتور","موس","کیبرد"]
for op in pc_part:
    listBox.insert(tk.END,op)

listBox.pack(padx=10,pady=10)

def delete_items():
    selecte = listBox.curselection()
    for i in reversed(selecte):
        listBox.delete(i)

ttk.Button(app,text="Delete",command=delete_items).pack()

add_entry = ttk.Entry(app)
add_entry.pack()

def add_item():
    var = add_entry.get()
    listBox.insert(tk.END,var)
    add_entry.delete(0,tk.END)

ttk.Button(app,text="Add",command=add_item).pack()

app.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk

app = tk.Tk()
app.geometry("300x400")

menu = tk.Menu(app)
app.config(menu=menu)

file_menu = tk.Menu(menu)
menu.add_cascade(label="File",menu =file_menu)
file_menu.add_command(label="open")
file_menu.add_command(label="Exit",command=app.quit)

tree = ttk.Treeview(app)

tree["col"] = ("c1","c2")

tree.heading("c1",text="Col 1")
tree.heading("c2",text="Col 2")

tree.insert("","end",text="time 1",values=("1.1","1.2"))
tree.insert("","end",text="time 2",values=("2.1","2.2"))

tree.pack()


listBox = tk.Listbox(app,selectmode=tk.MULTIPLE)
pc_part = ["مانیتور","موس","کیبرد"]
for op in pc_part:
    listBox.insert(tk.END,op)

listBox.pack(padx=10,pady=10)

def delete_items():
    selecte = listBox.curselection()
    for i in reversed(selecte):
        listBox.delete(i)

ttk.Button(app,text="Delete",command=delete_items).pack()

add_entry = ttk.Entry(app)
add_entry.pack()

def add_item():
    var = add_entry.get()
    listBox.insert(tk.END,var)
    add_entry.delete(0,tk.END)

ttk.Button(app,text="Add",command=add_item).pack()

app.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk

app = tk.Tk()
app.geometry("300x400")

main_fraim = tk.Frame(app)
main_fraim.pack(fill=tk.BOTH,expand=True)

text = tk.Text(main_fraim)
text.pack(fill=tk.BOTH,side=tk.LEFT,expand=True)

sc = tk.Scrollbar(main_fraim,command=text.yview)
sc.pack(side=tk.RIGHT,fill=tk.Y)
text.config(yscrollcommand=sc.set)

app.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk
from tkinter import filedialog
app = tk.Tk()
app.geometry("300x400")

menu = tk.Menu(app)
app.config(menu=menu)
text_w = tk.Text(app)
text_w.pack(fill="both",expand=True)

def open_file():
    global text_w
    file_path= filedialog.askopenfile()
    if file_path :
        with open(file_path.name,"r") as file :
            file_content  = file.read()
            text_w.delete(1.0,tk.END)
            text_w.insert(tk.END,file_content)

file_menu = tk.Menu(menu)
menu.add_cascade(label="File",menu =file_menu)
file_menu.add_command(label="New")
file_menu.add_command(label="Open",command=open_file)
file_menu.add_command(label="Save")
file_menu.add_separator()
file_menu.add_command(label="Exit",command=app.quit)


app.mainloop()
```
